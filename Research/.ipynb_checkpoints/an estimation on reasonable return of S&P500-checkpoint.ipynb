## calculate average EPS

```python
#drop negative stocks
data_filtered_dropNeg = data_filtered[(data_filtered['DEPS Y1'] > 0) & (data_filtered['DEPS Y1'] >0)]
```

```python
EPSGrowth = (data_filtered_dropNeg['DEPS Y1'] - data_filtered_dropNeg['DEPS Y2'])/data_filtered_dropNeg['DEPS Y2']
```

```python
average_EPSGrowth = EPSGrowth.mean()
```

```python
average_EPSGrowth
```

## calculate average inflation rate of US in 2017

method: subtract each month's CPI in 2017 with that of 2016. And average out the number. <br>
source: https://www.bls.gov/cpi/tables/supplemental-files/home.htm

```python
CPI_2017 = pd.DataFrame([242.839, 243.603,243.801,244.524,244.733,244.955,244.786,245.519,246.819,246.663,246.669,246.524])
```

```python
CPI_2016 = pd.DataFrame([236.916,237.111,238.132,239.261,240.229,241.018,240.628,240.849,241.428,241.729,241.353,241.432])
```

```python
CPI_percentChange = (CPI_2017 - CPI_2016)/CPI_2016
```

```python
average_CPIpercentChange = CPI_percentChange.mean()
```

```python
average_CPIpercentChange
```

## average dividend grwoth rate of SPX

source: http://www.multpl.com/s-p-500-dividend-growth

```python
average_divident = 0.0699
```

## Reasonable return of SPX

```python
reasonable_return_lower = average_EPSGrowth + average_divident + average_CPIpercentChange
print('The reasonable expected return is {}'.format(reasonable_return_lower))
```

```python
reasonable_return_higher = 0.1967 + average_divident + average_CPIpercentChange
print('The reasonable expected return is {}'.format(reasonable_return_higher))
```
